```python
# ===================================================
# Title: CRC-DHCP Simulation Visuals
# Purpose: Generate tables, flowchart, and bar graph
# Dependencies: matplotlib, pandas
# ===================================================

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches

print("Starting CRC-DHCP visual generation...")

# ---------------------------
# TABLE 1: DHCP CRC Mapping
# ---------------------------
data1 = {
    "Packet Type": ["DISCOVER", "OFFER", "REQUEST", "ACK"],
    "CRC Added": ["No", "Yes", "Yes", "Yes"],
    "Purpose": [
        "Client initiates lease request",
        "Server offers IP with CRC",
        "Client requests IP with CRC",
        "Server confirms lease with CRC"
    ]
}
df1 = pd.DataFrame(data1)
print("Table 1 Data:\n", df1)

fig1, ax1 = plt.subplots(figsize=(8, 2.5))
ax1.axis('off')
tbl1 = ax1.table(cellText=df1.values, colLabels=df1.columns, loc='center', cellLoc='center')
tbl1.auto_set_font_size(False)
tbl1.set_fontsize(10)
tbl1.scale(1.2, 1.5)
plt.tight_layout()
plt.savefig("Table1_DHCP_Packets_CRC.png", dpi=300)
plt.show()
print("✅ Table 1 saved: Table1_DHCP_Packets_CRC.png")

# ---------------------------
# TABLE 2: Performance Metrics
# ---------------------------
data2 = {
    "Metric": ["Integrity Detection", "IP Conflict Rate", "Lease Renewal Success", "Average Allocation Time"],
    "Standard DHCP": ["70%", "7.5%", "82%", "12 ms"],
    "Proposed CRC-DHCP": ["99%", "5.1%", "99%", "15 ms"]
}
df2 = pd.DataFrame(data2)
print("Table 2 Data:\n", df2)

fig2, ax2 = plt.subplots(figsize=(8, 2.5))
ax2.axis('off')
tbl2 = ax2.table(cellText=df2.values, colLabels=df2.columns, loc='center', cellLoc='center')
tbl2.auto_set_font_size(False)
tbl2.set_fontsize(10)
tbl2.scale(1.2, 1.5)
plt.tight_layout()
plt.savefig("Table2_DHCP_Performance_Comparison.png", dpi=300)
plt.show()
print("✅ Table 2 saved: Table2_DHCP_Performance_Comparison.png")

# ---------------------------
# FLOWCHART: DHCP + CRC Logic
# ---------------------------
print("Generating flowchart...")

fig3, ax3 = plt.subplots(figsize=(8, 6))
ax3.set_xlim(0, 10)
ax3.set_ylim(0, 12)
ax3.axis('off')

boxes = [
    (3.5, 10.5, 3, 1, "Client Sends\nDHCP DISCOVER"),
    (3.5, 9, 3, 1, "Server Sends\nOFFER + CRC"),
    (3.5, 7.5, 3, 1, "Client Verifies\nCRC of OFFER"),
    (3.5, 6, 3, 1, "Client Sends\nREQUEST + CRC"),
    (3.5, 4.5, 3, 1, "Server Verifies\nCRC of REQUEST"),
    (3.5, 3, 3, 1, "Server Sends ACK\n+ Optimized IP + CRC"),
    (3.5, 1.5, 3, 1, "Client Verifies\nCRC of ACK"),
    (3.5, 0, 3, 1, "IP Lease Confirmed")
]

# Draw boxes and text
for x, y, w, h, text in boxes:
    ax3.add_patch(patches.Rectangle((x, y), w, h, edgecolor='black', facecolor='lightblue'))
    ax3.text(x + w / 2, y + h / 2, text, ha='center', va='center', fontsize=9, fontweight='bold')

# Arrows
for i in range(len(boxes) - 1):
    x1 = boxes[i][0] + boxes[i][2] / 2
    y1 = boxes[i][1]
    x2 = boxes[i + 1][0] + boxes[i + 1][2] / 2
    y2 = boxes[i + 1][1] + boxes[i + 1][3]
    ax3.annotate("", xy=(x2, y2), xytext=(x1, y1), arrowprops=dict(arrowstyle="->", lw=1.5))

plt.tight_layout()
plt.savefig("CRC_DHCP_Flowchart.png", dpi=300)
plt.show()
print("✅ Flowchart saved: CRC_DHCP_Flowchart.png")

# ---------------------------
# RESULT METRICS BAR GRAPH
# ---------------------------
print("Generating comparison graph...")

metrics = ["Integrity Detection", "IP Conflict Rate", "Lease Renewal Success", "Avg. Allocation Time (ms)"]
standard = [70, 7.5, 82, 12]
proposed = [99, 5.1, 99, 15]
x = range(len(metrics))
bar_width = 0.35

fig4, ax4 = plt.subplots(figsize=(10, 6))
bars1 = ax4.bar([i - bar_width/2 for i in x], standard, bar_width, label='Standard DHCP', color='gray')
bars2 = ax4.bar([i + bar_width/2 for i in x], proposed, bar_width, label='Proposed CRC-DHCP', color='royalblue')

ax4.set_xlabel("Performance Metric")
ax4.set_ylabel("Value")
ax4.set_title("Comparison of DHCP and CRC-DHCP Performance")
ax4.set_xticks(x)
ax4.set_xticklabels(metrics, rotation=15)
ax4.legend()
ax4.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate values
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax4.annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                     xytext=(0, 3), textcoords="offset points", ha='center', fontsize=9)

plt.tight_layout()
plt.savefig("CRC_DHCP_Result_Comparison_Graph.png", dpi=300)
plt.show()
print("✅ Result graph saved: CRC_DHCP_Result_Comparison_Graph.png")

print("\n🎉 All visualizations generated successfully!")
```
